**GridSearch CV**

GridSearch CV is used to determining the best parameters

```python
#loading the SVC Model
model=SVC()
```

```python
#Hyperparameter
parameters={
          'kernel':['linear','poly','rbf','sigmoid'],
          'C':[1,5,10,15]
}
```

```python
#gridsearchcv
classifier=GridSearchCV(model,parameters,cv=5)
```

```python
#fitting the to our model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
#best parameter
best_parameter=classifier.best_params_
print(best_parameter)
```

```python
#best score
best_score=classifier.best_score_
print(best_score)
```

```python
#loading the result to pandas dataframe
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result=result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

High Accuracy=97.5\
best Parameter={'C': 5, 'kernel': 'rbf'}


**RandomizedSearchCV**

```python
model=SVC()
```

```python
#hyperparameter
parameters={'kernel':['linear','poly','rbf','sigmoid'],
            'C':[1,5,10,20]
            
    

}
```

```python
#random search
classifier=RandomizedSearchCV(model,parameters,cv=5)
```
